### Method and discussion:

Zakladamy tutaj nastepujace rownania Love'a:

$u_x = \frac{b_x}{2\pi}[arctg(\frac{y}{x}) + \frac{xy}{2(1-\nu)(x^2 + y^2)}]$

$u_y = -\frac{b_x}{2\pi}[\frac{1-2\nu}{4(1-\nu)}ln\frac{x^2+y^2}{r_o^2} + \frac{x^2-y^2}{4(1-\nu)(x^2+y^2)}]$

$u_z = \frac{bz}{2\pi}arctg(\frac{y}{x})$

Rozumiem, e istniej **zaoenia** co do tego, kiedy dane rwnanie moe by zaaplikowane, np. e kada paszczyzna krystalograficzna jest "paska".

Zaoenia te nie s spenione po zaaplikowaniu pierwszej dyslokacji. Moemy wic sprbowa przeksztaci siatk po pierwszej dyslokacji do nowej "przestrzeni", w ktrej te zaoenia bd juspenione.

Dla uproszczenia dalszej analizy przyjmijmy przez chwil, e warunek "paszczyzna krystalograficzna jest "paska"" oznacza, e wszystkie punkty na tej samej paszczynie krystalograficznej majtaksamwsprzdn$y$.

Moemy wic przyj nastpujcstrategi ($\mathbf{x} = (x, y, z)$, $\mathbf{u} = (u_x, u_y, u_z)$):

1. Przeksztacamy $\mathbf{x}$ do nowej przestrzeni: $\mathbf{x} \rightarrow \mathbf{x}'$, w ktrej wszystkie punkty z jednej paszczyzny maj takie same $y$.
2. W przestrzeni rozwizujemy rwnanie $\mathbf{u}' = f(\mathbf{x}'+\mathbf{u}')$ wzgl. $\mathbf{u}'$.
3. Przesuwamy o znalezione przesunicie $\mathbf{x}' \leftarrow \mathbf{x}' + \mathbf{u}'$.
4. Przeksztacamy w drug stron: $\mathbf{x}' \rightarrow \mathbf{x}$.

W jednej z prostszych sytuacji, jeeli chcemy speniwarunek e warunek "wszystkie punkty na tej samej paszczynie krystalograficznej majtaksamwsprzdn$y$", to moemy sprobwa przyj nastpujce przeksztacenie $\mathbf{x} \rightarrow \mathbf{x}'$.


$x' = x$

$y' = y - \Delta y(x, y, z)$

$z' = z$

oraz przeksztalcenie odwrotne:


$x = x'$

$y = y' + \Delta y(x, y, z)$

$z = z'$


Gdzie $\Delta y(x, y, z) = y_c - y_0$, gdzie $y_c$ to pooenie $y$ atomw z konfiguracji aktualnej, $y_0$ to pooenie atomw z konfiguracji pierwotnej (np. wyznaczone dla pierwszej z dyslokacji).


W punkcie 2. podstawiamy nowe $\mathbf{x}'$ do wzorw Love'a:

$u_x' = \frac{b_x}{2\pi}[arctg(\frac{y'}{x'}) + \frac{x'y'}{2(1-\nu)(x'^2 + y'^2)}]$

$u_y' = -\frac{b_x}{2\pi}[\frac{1-2\nu}{4(1-\nu)}ln\frac{x'^2+y'^2}{r_o^2} + \frac{x'^2-y'^2}{4(1-\nu)(x'^2+y'^2)}]$

$u_z' = \frac{bz}{2\pi}arctg(\frac{y'}{x'})$


(**UWAGA**: $y' = y - \Delta y(x, y, z)$, i naley brato bra pod uwag w sytuacji, gdyby bya potrzebna np. $\frac{\delta u_x'}{\delta y}$ (bety); my tu jednak zakadamy, e liczymy $\frac{\delta u_x'}{\delta y'}$)

Rozwizujemy rwnanie $\mathbf{u}' = f(\mathbf{x}'+\mathbf{u}')$ (np. metod Newtona-Raphsona), i przeksztacamy spowrotem do naszej oryginalnej przestrzeni uywajc przyjtej $\mathbf{x}' \rightarrow \mathbf{x}$.



## Initial configuration

```python
l0 = wzt.generate.create_lattice(
    dimensions=(10, 5, 2),  #  The number of cells
    cell="B4_ZnS",
)
```

```python
# wzt.visualization.render(molecule=l0, window_size=(800, 800))
```

```python
fig, ax = wzt.visualization.plot_atoms_2d(l0)
ax.set_title("Initial configuration")
```

## 1st dislocation

```python
b0 = [1, 0, 0]
position0 = [5, 6, 7.5]
plane0 = (0, 0, 1)

u0 = wzt.dislocations.displace_love(
    crystal=l0,
    position=position0,
    burgers_vector=b0,
    plane=plane0,
    bv_fraction=1.0,
)
```

```python
wzt.visualization.plot_displacement(l0, u0)
```

```python
bv = 1.0 * l0.cell.to_cartesian_indices(b0)
l1 = l0.translate(u0)
l1 = wzt.generate.update_bonds(l1)
fig, ax = wzt.visualization.plot_atoms_2d(l1)
wzt.visualization.plot_crystal_surface_y(ax, position0, x0=-0.2, y0=0.96, xlim=(-7, 25), color="red", linestyle="dashed", bx=bv[0])
```

## 2nd dislocation

```python
delta_y = np.zeros(l1.coordinates.shape)
delta_y[:, 1] = l1.coordinates[:, 1] - l0.coordinates[:, 1]
delta_y = np.squeeze(delta_y)
l1_t = l1.translate(-delta_y)
fig, ax = wzt.visualization.plot_atoms_2d(l1_t)
```
